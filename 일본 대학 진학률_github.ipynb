```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.rc("font", family = "Malgun Gothic")
sns.set(font="Malgun Gothic", 
rc={"axes.unicode_minus":False}, style='white')
import warnings
warnings.filterwarnings(action='ignore')
```

```python
df = pd.read_excel('D:/Data/project/Project_1st/01.미니프로젝트(8월)_기획안 및 일일양식/분석 데이터/jp/일본 대학 진학률_02.xlsx', index_col=0)
df = df[['고등학교 합계','대학교 합계']]
df
```

```python
fig = plt.figure(figsize=(10,5))
plt.style.use('ggplot')
ax = fig.add_subplot()
 
ax.plot(df.index,df['고등학교 합계'],marker='o',label='고등학교', linewidth=3.0, color = 'r', alpha = 0.8)
ax.plot(df.index,df['대학교 합계'],marker='o',label='대학교', linewidth=3.0, color = 'g', alpha = 0.8)
 
ax.legend()
plt.title('일본의 대학 진학률', fontsize=20)
plt.show()
```
